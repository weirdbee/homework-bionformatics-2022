```python
df_gng4 = pd.read_csv('GNG4_exons.tsv', sep = '\t', names = ['chr', 'smth', 'exon', 'start', 'end', '.', '-', '..1', '???'])
df_spr4 = pd.read_csv('SPRR4_exons.tsv', sep = '\t', names = ['chr', 'smth', 'exon', 'start', 'end', '.', '-', '..1', '???'])
```

```python
a = []
for i, j in zip(df_gng4['start'], df_gng4['end']):
  if (i, j) not in a:
    a.append((i, j))

b = [None] * (len(a) * 2)

for i in range(len(a)):
  b[i * 2] = (a[i][0], False)
  b[i * 2 + 1] = (a[i][1], True)
         
b = sorted(b, key = lambda x: x[0])
ans, count = 0, 0

for i in range(0, len(a) * 2):
  if (i != 0) & (b[i][0] > b[i - 1][0]) &  (count > 0):
    ans += b[i][0] - b[i - 1][0]
  if b[i][1]:
    count -= 1
  else:
    count += 1
print(f'len of GNG4 : {ans + 1}')
```

```python
a = []
for i, j in zip(df_spr4['start'], df_spr4['end']):
  if (i, j) not in a:
    a.append((i, j))

b = [None] * (len(a) * 2)

for i in range(len(a)):
  b[i * 2] = (a[i][0], False)
  b[i * 2 + 1] = (a[i][1], True)
         
b = sorted(b, key = lambda x: x[0])
ans, count = 0, 0

for i in range(0, len(a) * 2):
  if (i != 0) & (b[i][0] != b[i - 1][0]) &  (count > 0):
    ans += b[i][0] - b[i - 1][0]
  if b[i][1]:
    count -= 1
  else:
    count += 1
print(f'len of SPR4 : {ans + 1}')
```

```python
from scipy.stats import t
import numpy as np
```

```python
factors = [0.35219656, 0.39439086, 0.73057344, 1.66138079, 
1.60002838, 1.48313616, 1.28046971, 0.92434274, 1.59306799, 1.34997698]

df = pd.read_csv("TCGA-COAD_cancer_normal.tsv", sep="\t", index_col=0)
gl = pd.read_csv("gene_lengths.tsv", sep="\t", index_col=0).sort_index()
df.sort_index()
gl.sort_index()

RPM = df.div(df.sum(axis=0), axis = 1) * 1e+6
dsq2 = RPM.div(factors, axis = 1) 
dsq2 = RPM.div(gl["Length"], axis = 0) * 1000

dsq2 = dsq2[dsq2.min(axis = 1) > 0]

mean_tumor = np.mean(dsq2.iloc[:, :5], axis = 1)
mean_health = np.mean(dsq2.iloc[:, 5:], axis = 1)
std_tumor = dsq2.iloc[:, :5].std(axis = 1)
std_health = dsq2.iloc[:, 5:].std(axis = 1)

dsq2['t'] = abs((mean_tumor - mean_health)/(std_tumor**2/5 + std_health**2/5)**0.5)
dsq2 = dsq2[dsq2['t'] > abs(t.ppf(0.05/2, df =  8))]
dsq2 = dsq2.sort_values(by = 't', ascending = False)
dsq2.iloc[:10, [-1]]
```
